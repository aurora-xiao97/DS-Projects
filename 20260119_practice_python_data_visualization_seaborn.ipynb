## Introduction to Seaborn

Seaborn was built on top of matplotlib

to start with loading the package, you still need to load the maplotlib, since the Seaborn library was built on top of it.


### There are two types of plots 
 ** Relational plots
- Show the relationship between two quantitative variables
- Examples: scatter plots ("scatter"), line plots ("line")
 
 ** Categorical plots
- Show the distribution of a quantitative variable within categories defined by a categorical variable
- Examples: bar plots, count plots, box plots, point plots

adding a third variable (row/col)

setting row and/or col in relplot() or catplot() will create subgroups that are displayed on separate subplots




```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
heights = [165, 178, 182, 160, 175, 168, 190, 155, 172, 185, 162, 180, 158, 176, 170]
weights = [58, 75, 82, 52, 70, 63, 88, 48, 68, 80, 55, 77, 50, 73, 65]

df = pd.DataFrame({
    'height': heights,
    'weight': weights 

})
```

```python
## to draw a scatter plot
sns.scatterplot(data = df, x = 'height', y = 'weight')
plt.show()
```

```python
## example of bar plot

gender = ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F']
sns.countplot(x = gender)
plt.show()
```

### Scatter Plot

show relationship between two quantitative variables

```python
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
plt.show()
```

```python
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue='smoker') ## adding hue_order to modify the order of the legend
## adding hue_order to modify the order of the legend
plt.show()
```

```python
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue='smoker', hue_order=['Yes', 'No']) 
plt.show()
```

```python
## using dictionary to customize colors
hue_color = {"Yes": "black", "No": "red"}
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue='smoker', hue_order=['Yes', 'No'], palette=hue_color) 
plt.show()
```

### Subplots

#### Introducing relplot()
it allows you to create subplot in a single figure

```python
sns.relplot(data = tips, x = 'total_bill',  kind = "scatter", y = 'tip', col='smoker')
plt.show()
```

instead of showing two subplots side by side, you could change it to row

```python
sns.relplot(data = tips, x = 'total_bill', kind = "scatter", y = 'tip', row ='smoker')
plt.show()
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = "scatter", row ='smoker', col = 'time')
plt.show()
```

other useful parameters for the relplot() are:
col_wrap & col_order


### Subgroups with point size

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = "scatter", size = "size", hue = "size")
plt.show()
```

### Subgroups with point style

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = "scatter", hue = "smoker", style="smoker")
plt.show()
```

### Changing point transparency

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = "scatter", alpha=0.5)
plt.show()
```

## Count Plots and Bar Plots

### using catplot() to create categorical plots

```python
print(tips.head())
sns.countplot(data= tips, x = "time")
plt.show()
```

```python
print(tips.head())
sns.catplot(data= tips, x = "time", kind = "count")
plt.show()
```

### Changing the order

```python
### order parameter
time_order = ["Lunch", "Dinner"]
sns.catplot(data= tips, x = "time", kind = "count", order = time_order)
plt.show()
```

### Bar Plots

### Looking at the average bill price by day of the week

```python
sns.catplot(data= tips, x = "day", y = "total_bill", kind = "bar")
plt.show()
```

## Box Plot

#### Shows the distribution of quantitative data

```python
sns.catplot(data = tips, x = "time", y = "total_bill", kind = "box")
plt.show()
```

#### change the order of categories

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.catplot(data = tips, x = "time", y = "total_bill", kind = "box", order = ["Dinner", "Lunch"])
plt.show()
```

#### omitting the outliers using 'sym'

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.catplot(data=tips, x="time", y="total_bill", kind="box", sym="")
plt.show()
```

you could also modify the whis to

whis = 2.0 --> make them extend to 2.0* IQR: whis = 2.0

Show the 5th and 95th percentiles: whis = [5, 95]

show min and max values: whis = [0,100]

## Point plots vs. line plots

### what are the differences between point plots and line plots?
- Line plot has quantitative variable (usually time) on x-axis
- Point plot has categorical variable on x-axis

```python
sns.catplot(data= tips, x = "time", y = "total_bill", kind = "point", hue = "day", join = False, ci = None )
plt.show()
```

### FacetGrid vs. AxesSubplot objects
