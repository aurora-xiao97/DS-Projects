```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
mean = 5
lam = 1 / mean

# Define the x values (0 to 25 to show the decay)
x = np.linspace(0, 25, 1000)
# PDF formula: lambda * e^(-lambda * x)
y = lam * np.exp(-lam * x)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f'Exponential Distribution (Mean={mean})', color='blue', lw=2)
plt.fill_between(x, y, color='blue', alpha=0.1)

# Annotations
plt.title('Exponential Probability Density Function (Mean = 5)')
plt.xlabel('x (Value)')
plt.ylabel('f(x) (Density)')
plt.axvline(x=0, color='black', linewidth=1)
plt.axhline(y=0, color='black', linewidth=1)
plt.axvline(x=mean, color='red', linestyle='--', label=f'Mean = {mean}')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()

plt.show()
```
