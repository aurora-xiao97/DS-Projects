```python
## sorting the dataframe by singel column, release_year
netflix_data.sort_values(by = 'release_year', ascending = False)
```

```python
netflix_data.sort_values(by = ['release_year', 'duration'], ascending = [False, False])
## sorting the dataframe by multiple columns, release_year and duration
## ascending = [False, False] means both columns are sorted in descending order
```

```python
## return netflix_data released from 1990 to 1999
netflix_data[netflix_data['release_year'].between(1990, 1999)] 
## there is a parameter called inclusive, such as "neither", "both", "left", "right"
```

```python
netflix_data[(netflix_data['release_year']>=1990) & (netflix_data['release_year'] <= 1999)]
             ## another way to return netflix_data released from 1990 to 1999
```

```python
## Adding a new column to the dataframe called years_from_2025
netflix_data['years_from_2025'] = 2025 - netflix_data['release_year']
```
