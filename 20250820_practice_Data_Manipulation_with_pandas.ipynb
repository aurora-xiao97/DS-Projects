```python
## what is the mean, median, and mode, min, and max, var, and std, sum of the duration column
print("Mean:", netflix_data["duration"].mean())
print("Median:", netflix_data["duration"].median())
print("Mode:", netflix_data["duration"].mode())
print("Min:", netflix_data["duration"].min())
print("Max:", netflix_data["duration"].max())
print("Variance:", netflix_data["duration"].var())
print("Standard Deviation:", netflix_data["duration"].std())
print("Sum:", netflix_data["duration"].sum())
## what is the longest movie in the data set
longest_movie = netflix_data.loc[netflix_data["duration"].idxmax()]
print(netflix_data["duration"].idxmax())
print("The longest movie is:", longest_movie["title"], "with a duration of", longest_movie["duration"], "minutes, released in", longest_movie["release_year"])
```

### counting how many movie or tv show by country

```python
netflix_data.groupby(["country", "type"])["show_id"].agg('count').reset_index().sort_values(by ="show_id", ascending = False)
```

```python
## Simplify previous counting how many movie or tv show by country from this table

## index two columns country and type since we want to group by these two columns
## values is show_id since we want to count how many movie or tv show by country
## aggfunc is count since we want to count how many movie or tv show by country

netflix_data.pivot_table(index = ["country", "type"], values = "show_id", aggfunc = 'count').reset_index().sort_values(by ="show_id", ascending = False)
```

### Pivot Table

```python
country_type_duration = netflix_data.pivot_table(values = "duration", index = ["type", "country"], aggfunc = "sum").reset_index()
```

```python
country_type_duration.head()
```

```python
country_type_duration.groupby("type")["duration"].agg(["sum"])
```

```python
country_type_duration.groupby("type")["duration"].agg(["sum"]).plot(kind = "bar", title = "Total Duration by Type", xlabel = "Type", ylabel = "Total Duration (minutes)", color = ["skyblue", "lightgreen"])
plt.show()
```
