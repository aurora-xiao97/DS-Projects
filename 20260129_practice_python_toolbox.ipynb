Iterator:

### Iterators

#### Using enumerate()

enumerate is a function that takes any iterable as argument, such as a list, and returns a special enumerate object, consists of paris containing the elements of the original iterable, along with their index within the iterable

The enumerate object itself is also an iterable and we can loop over it while unpacking its elements using the clause for index, value in enumerate(avengers)

```python
avengers = ['hawkeye', 'iron man', 'thor', 'quicksilver']
```

```python
for index, value in enumerate(avengers):
    print(f"Index: {index}, Value: {value}")
```

#### Using zip()

```python
names = ['barton', 'stark', 'odinson', 'maximoff']
```

```python
for z1, z2 in zip(avengers, names):
    print(z1, z2)
```

#### Using zip(*) to unpack

```python
mutants = ('charles xavier',
 'bobby drake',
 'kurt wagner',
 'max eisenhardt',
 'kitty pryde')
```

```python
powers = ('telepathy',
 'thermokinesis',
 'teleportation',
 'magnetokinesis',
 'intangibility')
```

```python
# Create a zip object from mutants and powers: z1
z1 = zip(mutants, powers)

# Print the tuples in z1 by unpacking with *
print(*z1)

# Re-create a zip object from mutants and powers: z1
z1 = zip(mutants, powers)

# 'Unzip' the tuples in z1 by unpacking with * and zip(): result1, result2
result1, result2 = zip(*z1)

# Check if unpacked tuples are equivalent to original tuples
print(result1 == mutants)
print(result2 == powers)
```

### List Comprehensions

List comprehensions:

- Collapse for loops for building lists into a single line
- Components
    - Iterable
    - Iterator variable (represent members of iterable)
    - Output expression
    

```python
nums = [12, 8, 21, 3, 16]
new_nums = [num + 1 for num in nums]
```

```python
new_nums
```

#### creating a nested loops with traditional way

```python
pairs_1 = []
for num1 in range(0,2):
    for num2 in range(6,8):
        pairs_1.append((num1, num2))
print(pairs_1)
```

```python
pairs_2 = [(num1, num2) for num1 in range(0,2) for num2 in range(6,8)]
```

```python
pairs_2
```

```python
# Create a 5 x 5 matrix using a list of lists: matrix
matrix = [[col for col in range(5)] for row in range(5)]

# Print the matrix
for row in matrix:
    print(row)
```

#### Advanced comprehensions

```python
[num ** 2 for num in range(10) if num % 2 == 0]
```

```python
[num ** 2 if num % 2 == 0 else 0 for num in range(10)]
```

#### Dict comprehension

- create dictionaries
- Use curly braces {} instead of brackets []

```python
pos_neg = {num: -num for num in range(10)}
```
